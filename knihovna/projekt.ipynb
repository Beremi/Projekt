### Výpočet kovariance

```python
kov11 = moje_knihovna.kovariance(data1, data1, 240, 20)
kov12 = moje_knihovna.kovariance(data1, data2, 240, 20)
kov21 = moje_knihovna.kovariance(data2, data1, 240, 20)
kov22 = moje_knihovna.kovariance(data2, data2, 240, 20)
a = [kov11, kov12, kov21, kov22]
print(a)
print(len(a))
```

### Tvorba matice

```python
matice = moje_knihovna.tvorba_matice(a, 240, 20)
print(matice)
```

```python
matice_kov = moje_knihovna.kovariancni_matice(matice, 240, 20)
print(matice_kov)
```

```python
mat_inv = moje_knihovna.inverze(matice_kov)
print(mat_inv)
```

```python
c = moje_knihovna.c(mat_inv, 240, 20)
print(c)
```

```python
optvynos = moje_knihovna.optimalizace_vynos([mv1, mv2], c)
print(optvynos)
```

```python
optriziko = moje_knihovna.optimalizace_riziko(matice, c)
print(optriziko)
```

```python
d = moje_knihovna.d(mat_inv, [mv1, mv2])
print(d)
```

```python
di = moje_knihovna.di(mat_inv, [mv1, mv2])
print(di)
```

```python
par_vyn = moje_knihovna.par_vynos(optvynos, d, 0.07)
print(par_vyn)
```

```python
vahy_vyn = moje_knihovna.vahy(c, di, par_vyn)
print(vahy_vyn)
riziko = moje_knihovna.riziko_dan(vahy_vyn, matice)
print(riziko)
```

```python
par_riz = moje_knihovna.par_riz(optriziko, d, 0.049)
print(par_riz)
```

```python
vahy_riz = moje_knihovna.vahy(c, di, par_riz)
print(vahy_riz)
vynos = moje_knihovna.vynos_dan(vahy_riz, [stv1, stv2])
print(vynos)
```
